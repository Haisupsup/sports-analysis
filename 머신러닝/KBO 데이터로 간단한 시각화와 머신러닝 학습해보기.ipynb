```python
import pandas as pd
import statsmodels.api as sm

# 변수와 상수 추가
X = sm.add_constant(x_train)  # 상수항 추가
y = y_train

# 선택할 유의수준 설정
sl_remove = 0.05

# 초기 변수 리스트
variables = X.columns.tolist()

# 각 스텝별로 제거된 변수와 모델 정보 저장
steps = []
removed_variables = []
adj_r_squared_list = []

while len(variables) > 0:
    # 모델 적합
    model = sm.OLS(y, X[variables]).fit(disp=0)
    pvalues = model.pvalues[1:]  # 상수항 제외
    max_pval = pvalues.max()

    if max_pval > sl_remove:  # 가장 높은 p-value가 유의수준 초과
        remove_variable = pvalues.idxmax()
        variables.remove(remove_variable)  # 변수 제거
        removed_variables.append(remove_variable)
        adj_r_squared_list.append(model.rsquared_adj)  # 수정된 결정계수 저장
        steps.append(variables.copy())
    else:
        break
```
