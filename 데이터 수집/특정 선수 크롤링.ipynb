```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import time

driver_path = 'C://Users//user//jupyter//kbl 데이터 크롤링//chromedriver-win64//chromedriver-win64//chromedriver.exe'
service = Service(driver_path)
driver = webdriver.Chrome(service=service)

url = 'https://statiz.sporki.com/stats/?m=main&m2=pitching&m3=default&so=WAR&ob=DESC&year=2024&sy=&ey=&te=&po=&lt=10100&reg=R&pe=&ds=&de=&we=&hr=&ha=&ct=&st=&vp=&bo=&pt=&pp=&ii=&vc=&um=&oo=&rr=&sc=&bc=&ba=&li=&as=&ae=&pl=&gc=&lr=&pr=50&ph=&hs=&us=&na=&ls=&sf1=&sk1=&sv1=&sf2=&sk2=&sv2='
driver.get(url)

time.sleep(5)

# Locate the dropdown menu container
dropdown_container = WebDriverWait(driver, 20).until(
    EC.element_to_be_clickable((By.ID, 'select_reg'))
)

# Click the dropdown to display options
dropdown_container.click()

# Select the option with value 'A' from the dropdown
option = WebDriverWait(driver, 20).until(
    EC.element_to_be_clickable((By.XPATH, '//ul[@class="option_list"]/li[@class="option_item" and @value="A"]'))
)
option.click()

dropdown_container = WebDriverWait(driver, 20).until(
    EC.element_to_be_clickable((By.ID, 'select_team'))
)

# Click the dropdown to display options
dropdown_container.click()

option = WebDriverWait(driver, 20).until(
    EC.element_to_be_clickable((By.XPATH, '//ul[@class="option_list"]/li[@class="option_item" and @value="1001"]'))
)
option.click()

# Continue with the rest of your code to interact with the page
# Locate and click the search button
#search_button = WebDriverWait(driver, 20).until(
#    EC.element_to_be_clickable((By.XPATH, "//button[@onclick='searchStats();']"))
#)
#search_button.click()

#time.sleep(3)

soup = BeautifulSoup(driver.page_source, 'html.parser')

# Locate the table containing the desired data
table_div = soup.find('div', class_='table_type01 transverse_scroll cbox')
if table_div:
    table = table_div.find('table')
    tbody = table.find('tbody')

    # Extract table rows
    rows = tbody.find_all('tr')
    all_data = []
    for row in rows:
        cells = row.find_all('td')
        row_data = [cell.get_text(strip=True) for cell in cells]
        all_data.append(row_data)

    # Create a DataFrame from the extracted data
    df = pd.DataFrame(all_data)

    # Set column headers if available
    thead = table.find('thead')
    if thead:
        headers = [th.get_text(strip=True) for th in thead.find_all('th')]
        df.columns = headers

    # Display the first few rows of the DataFrame
    print(df.head())

    # Save the DataFrame to a CSV file
    df.to_csv('kia_all_players.csv', encoding='utf-8-sig', index=False)
else:
    print('Unable to locate the data table. Please verify the HTML structure.')

driver.quit()
```
