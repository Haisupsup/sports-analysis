```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import time

# WebDriver 설정
driver_path = "C://Users//user//jupyter//kbl 데이터 크롤링//chromedriver-win64//chromedriver-win64//chromedriver.exe"
service = Service(driver_path)
driver = webdriver.Chrome(service=service)

# 사이트 접속
url = "https://statiz.sporki.com/stats/?m=main&m2=batting&m3=default&so=WAR&ob=DESC&year=2024&sy=&ey=&te=&po=&lt=10100&reg=R&pe=&ds=&de=&we=&hr=&ha=&ct=&st=&vp=&bo=&pt=&pp=&ii=&vc=&um=&oo=&rr=&sc=&bc=&ba=&li=&as=&ae=&pl=&gc=&lr=&pr=50&ph=&hs=&us=&na=&ls=&sf1=&sk1=&sv1=&sf2=&sk2=&sv2="
driver.get(url)

# 페이지 로드 대기
time.sleep(3)

# HTML 파싱
soup = BeautifulSoup(driver.page_source, "html.parser")

# 테이블 추출
table_div = soup.find("div", class_="table_type01 transverse_scroll cbox")
if table_div:
    table = table_div.find("table")
    tbody = table.find("tbody")

    # 테이블 데이터 추출
    rows = tbody.find_all("tr")
    all_data = []
    for row in rows:
        cells = row.find_all("td")
        row_data = [cell.text.strip() if cell else "" for cell in cells]
        all_data.append(row_data)

    # 데이터프레임 생성
    df = pd.DataFrame(all_data)

    # 컬럼명 설정 (헤더 추출)
    thead = table.find("thead")
    if thead:
        headers = [th.text.strip() for th in thead.find_all("th")]
        if len(headers) == df.shape[1]:
            df.columns = headers
        else:
            df.columns = [f"Column_{i}" for i in range(df.shape[1])]

    # 결과 확인
    print(df.head())

    # CSV 파일로 저장
    df.to_csv("statiz_batting_data.csv", encoding="utf-8-sig", index=False)
else:
    print("테이블을 찾을 수 없습니다. HTML 구조를 확인하세요.")

# WebDriver 종료
driver.quit()
```
