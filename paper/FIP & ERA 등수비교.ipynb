```python
another_sp_fip = data[['FIP', 'xFIP']]
```

```python
from sklearn.preprocessing import StandardScaler
from scipy.stats import pearsonr
from sklearn.metrics import mean_squared_error
import numpy as np

# 데이터 표준화
scaler = StandardScaler()
scaled_data = scaler.fit_transform(another_sp_fip)

# 표준화된 데이터를 다시 데이터프레임으로 변환
scaled_df = pd.DataFrame(scaled_data, columns=another_sp_fip.columns)

# Pearson 상관계수 및 p-value 계산
corr, p_value = pearsonr(scaled_df['FIP'], scaled_df['xFIP'])
print(f"\n### 상관계수 및 유의계수 ###")
print(f"Pearson Correlation Coefficient: {corr}")
print(f"P-value: {p_value}")

# RMSE 계산
rmse = np.sqrt(mean_squared_error(scaled_df['FIP'], scaled_df['xFIP']))
print(f"RMSE: {rmse}")

from sklearn.metrics import r2_score

r2 = r2_score(scaled_df['FIP'], scaled_df['xFIP'])

print(f"\n### 결정계수 (R²) ###")
print(f"R² Score: {r2}")
```
