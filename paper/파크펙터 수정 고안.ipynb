```python
import statsmodels.api as sm
import pandas as pd
import itertools

# 분석 대상 변수 설정

# 종속 변수 간 교호작용 변수 생성
interaction_terms = ['{}:{}'.format(a, b) for a, b in itertools.combinations(dependent_vars, 2)]  # 두 개씩 조합

# 결과 저장 리스트
results = []

# 다중 회귀 분석 수행
for target in dependent_vars:
    X = data[independent_vars].copy()  # 독립 변수 (BBF, PPF 등)
    y = data[target]  # 종속 변수 (ERA, FIP, WHIP 등)

    # 종속 변수들 간의 교호작용 추가
    for term in interaction_terms:
        var1, var2 = term.split(':')
        X[term] = data[var1] * data[var2]

    X = sm.add_constant(X)  # 상수항 추가
    model = sm.OLS(y, X).fit()  # OLS 회귀모델 적합

    # 각 독립 변수 및 종속 변수 교호작용 변수의 회귀 계수와 p-value 저장
    for var in independent_vars + interaction_terms:
        results.append((target, var, model.params[var], model.pvalues[var]))

# 데이터프레임 변환 및 출력
regression_results = pd.DataFrame(results, columns=['Target', 'Predictor', 'Coefficient', 'P-value'])
print(regression_results)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

# 예시: 4개 사이트의 파크팩터 리스트
park_factors = {
    "A_site": ppf1,
    "B_site": ppf2,
    "C_site": ppf3,
}

results = []

for name, ppf in park_factors.items():
    X_train, X_test, Y_train, Y_test = train_test_split(data1, ppf1, test_size=0.2, random_state=2025)
    reg = LinearRegression().fit(X_train, Y_train)
    y_pred = reg.predict(X_test)
    
    r2 = r2_score(Y_test, y_pred)
    
    n = X_test.shape[0]
    p = X_test.shape[1]
    r2_adj = 1 - ((1 - r2) * (n - 1) / (n - p - 1))

    rmse = np.sqrt(mean_squared_error(Y_test, y_pred))

    results.append({
        "Site": name,
        "R_squared": r2,
        "Adjusted_R_squared": r2_adj,
        "RMSE": rmse
    })
    
# 결과를 DataFrame으로 정리
results_df = pd.DataFrame(results)

# 📊 시각화 (막대그래프)
fig, axes = plt.subplots(1, 3, figsize=(30, 10))

# R²
axes[0].bar(results_df["Site"], results_df["R_squared"], color='skyblue')
axes[0].set_title("R² by Park Factor")
axes[0].set_ylabel("R²")
axes[0].set_ylim(0, 1)

# Adjusted R²
axes[1].bar(results_df["Site"], results_df["Adjusted_R_squared"], color='orange')
axes[1].set_title("Adjusted R² by Park Factor")
axes[1].set_ylabel("Adjusted R²")
axes[1].set_ylim(0, 1)

# RMSE
axes[2].bar(results_df["Site"], results_df["RMSE"], color='green')
axes[2].set_title("RMSE by Park Factor")
axes[2].set_ylabel("RMSE")
axes[2].invert_yaxis()  # 낮을수록 좋은 지표이므로 시각적으로 강조

plt.tight_layout()
plt.show()
```
