```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# 예제용 데이터 생성
X, y = make_classification(n_samples=1000, n_classes=2, random_state=42)

# 학습/테스트 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 분류기 모델 학습
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# 예측 확률(positive 클래스 확률) 계산
y_scores = clf.predict_proba(X_test)[:, 1]

# ROC curve 계산
fpr, tpr, thresholds = roc_curve(y_test, y_scores)
roc_auc = auc(fpr, tpr)

# ROC curve 시각화
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='orange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')  # 대각선 기준선
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.show()
```

```python
import pandas as pd

# 가중치 설정 (6~9회)
inning_weights = {
    6: 0.622,
    7: 0.203,
    8: 0.034,
    9: 0.005
}

# Modified ERA 계산 함수
def calculate_weighted_modified_era(row):
    # 1~5회 총 ER, IP 계산
    ER_early = sum([row[f"{i}ER"] for i in range(1, 6)])
    IP_early = sum([row[f"{i}IP"] for i in range(1, 6)])
    
    # ERA_early 계산
    ERA_early = (ER_early * 9 / IP_early) if IP_early > 0 else 0

    # 총 IP
    IP_total = sum([row[f"{i}IP"] for i in range(1, 10)])

    # 후반부 이닝별 가중 ERA 계산
    weighted_late_eras = []
    for i in range(6, 10):
        ER_i = row[f"{i}ER"]
        IP_i = row[f"{i}IP"]
        wi = inning_weights[i]

        if IP_i > 0:
            ERA_i = (ER_i * 9) / IP_i
            contribution = (IP_i / IP_total) * wi * ERA_i
            weighted_late_eras.append(contribution)

    ERA_late = sum(weighted_late_eras)

    # 전체 Modified ERA (가중 평균 방식)
    Modified_ERA = (IP_early / IP_total) * ERA_early + ERA_late
    return Modified_ERA

# 예시 데이터프레임 df에 적용
data["Modified_ERA_weighted"] = data.apply(calculate_weighted_modified_era, axis=1)
```

```python
import numpy as np

# 기존 이닝별 raw 가중치 (예시: 6~9회)
raw_weights = np.array([0.622, 0.203, 0.034, 0.005])

# 1. Inverse Frequency Weighting (희귀 이닝에 가중치 증가)
epsilon = 1e-5  # 0으로 나누는 오류 방지용 작은 수
inv_freq_weights = 1 / (raw_weights + epsilon)

# 2. Logistic Scaling 함수 적용
def logistic_scaling(x, threshold=1.5, scale=5):
    """
    Logistic 함수로 가중치 스케일링.
    threshold: logistic 곡선 중앙 위치 (조절 가능)
    scale: logistic 곡선 기울기 (조절 가능)
    """
    return 1 / (1 + np.exp(-scale * (x - threshold)))

# Logistic 함수에 인풋으로 inv_freq_weights의 로그값 사용 (log scale은 왜곡 완화용)
log_input = np.log(inv_freq_weights)
scaled_weights = logistic_scaling(log_input)

# 3. 정규화
final_weights = scaled_weights / scaled_weights.sum()

print("Raw weights:", raw_weights)
print("Inverse frequency weights:", inv_freq_weights)
print("Log input (log of inv freq):", log_input)
print("Scaled weights (after logistic):", scaled_weights)
print("Final normalized weights:", final_weights)
```

```python
import numpy as np
import scipy.stats as stats
import math

# 예시: 평균 이닝당 실점 (λ)
lambda_val = 0.85  # 이닝당 평균 실점

# 계산할 실점 수 k (0 ~ 5회까지 예시)
k_values = np.arange(0, 6)

# 1. 직접 수식으로 계산
def poisson_pmf(k, lam):
    return (lam**k * np.exp(-lam)) / math.factorial(k)

print("Poisson PMF using formula:")
for k in k_values:
    print(f"P(X={k}) = {poisson_pmf(k, lambda_val):.4f}")

# 2. scipy로 계산
print("\nPoisson PMF using scipy.stats.poisson:")
for k in k_values:
    print(f"P(X={k}) = {stats.poisson.pmf(k, mu=lambda_val):.4f}")
```

```python
# 이닝별 가중치
era_weights = {
    6: 1.475,
    7: 2.513,
    8: 4.61,
    9: 6.22
}

def compute_modified_era(row):
    # 1~5회
    er_early = sum([row[f'{i}ER'] for i in range(1, 6)])
    ip_early = sum([row[f'{i}IP'] for i in range(1, 6)])
    era_early = (er_early * 9 / ip_early) if ip_early > 0 else 0

    # 6~9회 (가중 조정)
    er_late = sum([row[f'{i}ER'] for i in range(6, 10)])
    ip_late = sum([row[f'{i}IP'] for i in range(6, 10)])
    gip_late = sum([row[f'{i}IP'] * era_weights[i] for i in range(6, 10)])
    era_late = (er_late * 9 / gip_late) if gip_late > 0 else 0

    ip_total = ip_early + ip_late
    if ip_total == 0:
        return None  # or np.nan

    # 가중 평균 ERA 계산
    modified_era = (era_early * (ip_early / ip_total)) + (era_late * (ip_late / ip_total))
    return modified_era

# 적용
data['Modified_ERA'] = data.apply(compute_modified_era, axis=1)
```

```python
# 이닝별 가중치 설정
a = np.log10(1/0.631)
b = np.log10(1/0.216)
c = np.log10(1/0.035)
d = np.log10(1/0.005)
print(a,b,c,d)
era_weights = {
    6: 1 + a,
    7: 1 + b,
    8: 1 + c,
    9: 1 + d
}

def compute_fully_adjusted_era(row):
    # 전체 실점 계산
    total_er = sum([row[f'{i}ER'] for i in range(1, 10)])

    # 1~5회 이닝 합계
    ip_early = sum([row[f'{i}IP'] for i in range(1, 6)])

    # 6~9회: 가중 이닝 계산
    weighted_ip_late = sum([row[f'{i}IP'] * era_weights[i] for i in range(6, 10)])

    # 가중 이닝 총합
    ip_adjusted_total = ip_early + weighted_ip_late

    # ERA 계산
    if ip_adjusted_total == 0:
        return None
    else:
        return (total_er * 9) / ip_adjusted_total

# 적용
data['Adjusted_ERA'] = data.apply(compute_fully_adjusted_era, axis=1)
data
```

```python
# 이닝별 가중치 설정
a = np.log10(1/0.631)
b = np.log10(1/0.216)
c = np.log10(1/0.035)
d = np.log10(1/0.005)
print(a,b,c,d)

era_weights = {
    6: 1 + a,
    7: 1 + b,
    8: 1 + c,
    9: 1 + d
}

def compute_fully_adjusted_era(row):
    # 전체 실점 계산
    total_er = sum([row[f'{i}ER'] for i in range(1, 10)])

    # 1~5회 이닝 합계
    ip_early = sum([row[f'{i}IP'] for i in range(1, 6)])

    # 6~9회: 가중 이닝 계산
    weighted_ip_late = sum([row[f'{i}IP'] * era_weights[i] for i in range(6, 10)])

    # 가중 이닝 총합
    ip_adjusted_total = ip_early + weighted_ip_late

    # ERA 계산
    if ip_adjusted_total == 0:
        return None
    else:
        return (total_er * 9) / ip_adjusted_total

# 적용
data['ai_ERA'] = data.apply(compute_fully_adjusted_era, axis=1)
def compute_adjusted_ip(row):
    ip_early = sum([row[f'{i}IP'] for i in range(1, 6)])
    weighted_ip_late = sum([row[f'{i}IP'] * era_weights[i] for i in range(6, 10)])
    return ip_early + weighted_ip_late

data['Adjusted_IP'] = data.apply(compute_adjusted_ip, axis=1)
data
```

```python
data['dif'] = data['ERA'] - data['ai_ERA']
data[data['dif']  < 0]
```

```python
def compute_era_minus(row):
    if row['Season_Adjusted_ERA'] == 0:
        return None
    park_adjusted_era = row['ai_ERA'] + (row['ai_ERA'] - (row['ai_ERA'] * row['A_PPF']))
    return 100 * (park_adjusted_era / row['Season_Adjusted_ERA'])

# 적용
data['aiERA'] = data.apply(compute_era_minus, axis=1)
data['aiERA']
```

```python
import pandas as pd
from scipy.stats import pearsonr
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

metrics = ['aiERA-', 'FIP-', 'ERA-', 'WAR']

# (1) 두 시즌 이상 뛴 선수만 필터링
player_counts = data['Name'].value_counts()
multi_season_players = player_counts[player_counts > 1].index
filtered = data[data['Name'].isin(multi_season_players)]

# (2) 시즌별 정렬
filtered = filtered.sort_values(by=['Name', 'Season']).reset_index(drop=True)

# (3) 그래프 및 상관 분석 반복
plt.figure(figsize=(16, 12))
results = []

for i, metric in enumerate(metrics, 1):
    # 다음 시즌 데이터 병합
    next_season = filtered.copy()
    next_season['Season'] = next_season['Season'] - 1
    merged = pd.merge(
        filtered,
        next_season[['Name', 'Season', metric]],
        on=['Name', 'Season'],
        suffixes=('', '_next')
    )
    
    valid = merged.dropna(subset=[metric, f"{metric}_next"])
    x = valid[metric]
    y = valid[f"{metric}_next"]

    # 상관 분석
    r, p = pearsonr(x, y)
    n = len(valid)
    se = np.sqrt((1 - r**2) / (n - 2)) if n > 2 else np.nan  # 표준오차 계산
    results.append((metric, r, p, se))

    # 회귀 시각화
    plt.subplot(2, 2, i)
    sns.regplot(
        x=x,
        y=y,
        ci=95,  # 95% 신뢰구간
        scatter_kws={'color': 'red', 'alpha': 0.6, 's': 30},
        line_kws={'color': 'blue'}
    )
    plt.xlabel(f"Season {metric}")
    plt.ylabel(f"Next Season {metric}")
    plt.title(f"{metric} Correlation: r={r:.3f}")
    plt.grid(True)

plt.tight_layout()
plt.show()

# (4) 결과 출력
print("📊 Correlation Summary (with Standard Error)")
for metric, r, p, se in results:
    signif = "✅" if p < 0.05 else "⚠️"
    print(f"{metric:7s}: r = {r:.4f}, p = {p:.4f}, SE = {se:.4f} {signif}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.utils import resample
import matplotlib.pyplot as plt
import numpy as np

# ROC Curve 시각화
plt.figure(figsize=(8, 6))

metrics = ['WAR', 'ERA-', 'FIP-', 'aiERA-']
colors = ['blue', 'green', 'purple', 'red']
line_styles = ['--', '-.', ':', '-']  # WAR: dashed, aiERA: solid

n_bootstrap = 500
mean_fpr = np.linspace(0, 1, 100)

for col, color, style in zip(metrics, colors, line_styles):
    model = LogisticRegression()
    model.fit(X_train[[col]], y_train)
    
    tpr_list = []
    auc_list = []

    # Bootstrap Resampling
    for i in range(n_bootstrap):
        X_bs, y_bs = resample(X_test[[col]], y_test, replace=True, random_state=i)
        y_prob = model.predict_proba(X_bs)[:, 1]
        fpr, tpr, _ = roc_curve(y_bs, y_prob)
        tpr_interp = np.interp(mean_fpr, fpr, tpr)
        tpr_interp[0] = 0.0  # TPR은 0부터 시작해야 함
        tpr_list.append(tpr_interp)
        auc_list.append(roc_auc_score(y_bs, y_prob))

    # 평균 TPR과 AUC 계산
    mean_tpr = np.mean(tpr_list, axis=0)
    mean_tpr[-1] = 1.0  # 마지막은 1로
    mean_auc = np.mean(auc_list)

    # ROC 곡선 그리기
    plt.plot(mean_fpr, mean_tpr, linestyle=style, color=color,
             label=f"{col} (AUC = {mean_auc:.3f})")

# 무작위 기준선
plt.plot([0, 1], [0, 1], linestyle=':', color='gray', label='Random (AUC = 0.5)')

plt.title("Smoothed ROC Curve with Bootstrap + Interpolation")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="lower right")
plt.grid(False)
plt.tight_layout()
plt.show()

c_index_dict = {}

for col in metrics:
    model = LogisticRegression()
    model.fit(X_train[[col]], y_train)
    y_prob = model.predict_proba(X_test[[col]])[:, 1]
    c_index = roc_auc_score(y_test, y_prob)  # C-index = ROC AUC
    c_index_dict[col] = c_index

# 출력
print("\n📊 C-index (ROC AUC와 동일):")
for col, ci in c_index_dict.items():
    print(f"{col:6s}: C-index = {ci:.4f}")
```

```python
import numpy as np

# 이닝별 가중치 정의
a = np.log10(74.4/47)
b = np.log10(74/16.1)
c = np.log10(74/2.6)
d = np.log10(74/0.4)

era_weights = {
    6: 1 + a,
    7: 1 + b,
    8: 1 + c,
    9: 1 + d
}

# 연도별 계산을 위한 초기 딕셔너리
year_stats = {}

# 연도별로 순회
for year, group in data.groupby('Year'):
    total_er = 0
    total_adjusted_ip = 0
    
    for _, row in group.iterrows():
        # 전체 실점
        er = sum([row[f'{i}ER'] for i in range(1, 10)])
        
        # 1~5회: 원래 이닝
        ip_early = sum([row[f'{i}IP'] for i in range(1, 6)])
        
        # 6~9회: 가중 이닝
        weighted_ip_late = sum([row[f'{i}IP'] * era_weights[i] for i in range(6, 10)])
        
        # 전체 조정 이닝
        adjusted_ip = ip_early + weighted_ip_late
        
        total_er += er
        total_adjusted_ip += adjusted_ip

    # ERA 계산
    adjusted_era = (total_er * 9) / total_adjusted_ip if total_adjusted_ip > 0 else None
    year_stats[year] = adjusted_era

# 결과 보기
for y in sorted(year_stats):
    print(f"{y}년 Adjusted ERA: {year_stats[y]:.3f}")
```

```python
# 이닝별 가중치
era_weights = {
    6: 1.2,
    7: 1.5,
    8: 2.0,
    9: 3.0
}

def compute_modified_era(row):
    # 1~5회
    er_early = sum([row[f'{i}ER'] for i in range(1, 6)])
    ip_early = sum([row[f'{i}IP'] for i in range(1, 6)])
    era_early = (er_early * 9 / ip_early) if ip_early > 0 else 0

    # 6~9회
    ip_late = sum([row[f'{i}IP'] for i in range(6, 10)])
    er_late = sum([row[f'{i}ER'] for i in range(6, 10)])
    gip_late = sum([row[f'{i}IP'] * era_weights[i] for i in range(6, 10)])

    # 수정된 후반 ERA
    if gip_late > 0:
        era_late = (er_late * 9 / gip_late) * (ip_late / gip_late)
    else:
        era_late = 0

    ip_total = ip_early + ip_late
    if ip_total == 0:
        return None  # or np.nan

    # 최종 수정 ERA
    modified_era = (era_early * (ip_early / ip_total)) + (era_late * (ip_late / ip_total))
    return modified_era
data['Modified_ERA'] = data.apply(compute_modified_era, axis=1)
data
```
